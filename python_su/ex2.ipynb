```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
```

```python
a = 50.0 * u.meter
print(a)
b = [23,45,88] * u.meter

c = [4,7,10] * u.second
print(c)
```

```python
plt.figure()
plt.plot(b,c,'o', label='Wind speed [m/s]')
plt.xlabel('Distance [m]')
plt.ylabel('Time [s]')
plt.legend()
```
